## Q1 Login to your Instagram Handle

```python
#importing relevant libraries
from selenium import webdriver
import time
import re
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import TimeoutException
```

```python
#opening chromedriver 
driver = webdriver.Chrome("C:\\chromedriver.exe")

#maximize the window
driver.maximize_window()
```

```python
def insta_openLoginPage():
    """
        Objective:
            This function is used to open the instagram login page.
        Input:
            None
        Output:
            IG login page opens up.
    """
    #sign up page opens up
    driver.get("https://www.instagram.com/")
    
    #wait until login link visible
    wait=WebDriverWait(driver,10) 
    login= wait.until(EC.visibility_of_element_located((By.CLASS_NAME,'izU2O')))
    
    #path for login page is extracted
    x=login.get_attribute('innerHTML').split('\"')
    path='https://www.instagram.com'+x[1]
    driver.get(path)

def dismis_notification():
    """
        Objective:
            This function is used to dismiss if notification box pops up .
        Input:
            None
        Output:
            notification box is dismissed.
    """
    try:
        #wait until notification box is located
        wait=WebDriverWait(driver,10) 
        notif_box= wait.until(EC.presence_of_element_located((By.XPATH,'//div[(@class="mt3GC")]/button[2]')))
        notif_box.click()
    except TimeoutException:
        pass
    
def insta_login(enter_username,enter_password):
    """
        Objective:
            This function is used for instagram login.
        Input:
            enter_username:string
            enter_password:string
        Output:
            IG User is logged in.
    """
    insta_openLoginPage()
    
    #enter username
    #wait until unsername text box is located
    wait=WebDriverWait(driver,10) 
    username= wait.until(EC.visibility_of_element_located((By.NAME,"username")))
    username.send_keys(enter_username)
    
    #enter password
    password=driver.find_element_by_name("password")
    password.send_keys(enter_password)
    
    #click on submit
    submit=driver.find_element_by_xpath('//button[(@type="submit")]')
    submit.click()
    
    #if notification box pops up 
    dismis_notification()
```

## Q2 Type for “food” in search bar and print all the names of the Instagram Handles that are displayed in list after typing “food”
Note : Make sure to avoid printing hashtags

```python
def insta_search(search_item):
    """
        Objective:
            This function is used for searching search_item using search bar.
        Input:
            search_item:string
        Output:
            IG User gets list of items options displayed on screen after searching search_item.
    """
    #if valid username or id is entered
    try:
        search=driver.find_element_by_xpath('//input[(@placeholder="Search")]')
        search.get_attribute('outerHTML')
        search.send_keys(search_item)
        search.send_keys(Keys.ENTER)
        
    #if invalid username or id is entered
    except NoSuchElementException:
        username=input('Enter Login ID: ')
        password=input('Enter Login PASSWORD: ')
        insta_login(username,password)        
        printName_IGHandles('food')
        
def clear_search_bar():
    """
        Objective:
            This function is used for clearing search bar.
        Input:
            none
        Output:
            search tab is cleared.
    """
    #wait until searh bar is located
    wait=WebDriverWait(driver,10) 
    cross= wait.until(EC.presence_of_element_located((By.XPATH,'//div[(@class="aIYm8 coreSpriteSearchClear")]')))
    cross.get_attribute('outerHTML')
    cross.click()
    
def Name_IGHandles(search_item):
    """
        Objective:
            This function is used for creating a list of IG Handle names after searching search_item.
        Input:
            search_item:string
        Output:
            Returns the list of IG Handle names(excludings hashtags).
    """
    insta_search(search_item)
    #wait until list generated after search
    wait=WebDriverWait(driver,20) 
    lst_search= wait.until(EC.presence_of_element_located((By.XPATH,'//div[(@class="fuqBx")]')))
        
    #extracting list generated after search                 
    lst_search=lst_search.find_elements_by_css_selector('a')
    
    #extracting IG handles(accounts) excluding location and tags
    IGHandle_Name=[]
    for i in lst_search:
        url=i.get_attribute('href')
        if '/explore/' not in url:
            IGHandle_Name.append(url.split('/')[3])
            
    #clear search bar
    clear_search_bar()
    
    #return IGHandle Name
    return IGHandle_Name

def printName_IGHandles(IGHandle_Name):
    """
        Objective:
            This function is used for printing list of IG Handle names after searching search_item.
        Input:
            IGHandle_Name:list
        Output:
            Print the list of IG Handle names(excludings hashtags).
    """
    #printing the list of IGHandle Name
    for i in IGHandle_Name:
            print(i)
```

## Q3 Searching and Opening a profile using 
Open profile of “So Delhi” 

```python
def insta_openProfile(search_item):
    """
        Objective:
            This function is used for opening the profile of search_item.
        Input:
            search_item:sting
        Output:
            search_item profile opens up.
    """
    IGHandle_Name=Name_IGHandles(search_item)
    
    #coverting search_item to lowercase and removing space
    search_item=search_item.lower().replace(' ','')
    
    #creating base url
    base_url='https://www.instagram.com/'
    
    #extracting username for current IG handle
    current_pagename=driver.current_url.split('/')[3]
    
    #if we are already on same page that is to be searched
    if current_pagename==search_item:
        pass
    else:
        for i in IGHandle_Name:
            if i==search_item:
                url=base_url+i
                driver.get(url)
                break
```

## Q4 Follow/Unfollow given handle - 
Open the Instagram Handle of “So Delhi”
Start following it. Print a message if you are already following
After following, unfollow the instagram handle. Print a message if you have already unfollowed.

```python
def insta_follow(search_item):
    """
        Objective:
            This function is used for following search_item IG page.
        Input:
            search_item:sting
        Output:
            search_item IG page is now showing following.
    """
    insta_openProfile(search_item)
    
    #wait until follow button is clickable
    wait=WebDriverWait(driver,10) 
    follow= wait.until(EC.element_to_be_clickable((By.XPATH,'//section[(@class="zwlfE")]/div/div/span/span/button'))) 
    
    #check if already following
    if 'Following' == follow.text:
            print('Already following')
    else:
        follow.click()
```

```python
def insta_unfollow(search_item):
    """
        Objective:
            This function is used for unfollowing search_item IG page.
        Input:
            search_item:sting
        Output:
            search_item IG page is now showing follow.
    """
    insta_openProfile(search_item)
    #wait until unfollow button is clickable
    wait=WebDriverWait(driver,10) 
    unfollow= wait.until(EC.element_to_be_clickable((By.XPATH,'//section[(@class="zwlfE")]/div/div/span/span/button')))
    
    #check if already unfollowing
    if 'Follow' == unfollow.text:
            print('Already unfollowed')
    else:
        unfollow.click()
        #confirmation box pops up
        wait=WebDriverWait(driver,10) 
        unfollow= wait.until(EC.element_to_be_clickable((By.XPATH,'//div[(@class="mt3GC")]/button')))  
        unfollow.click()
```

## Q5 Like/Unlike posts
Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.
Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.

```python
def click_right_arrow():
    """
        Objective:
            This function is used for clicking right arrow for next post.
        Input:
            None
        Output:
            next post opens up.
    """
    driver.find_element_by_class_name('coreSpriteRightPaginationArrow').click()
    
def click_close_post():
    """
        Objective:
            This function is used for closing current post.
        Input:
            None
        Output:
            current post is closed.
    """
    driver.find_element_by_class_name("ckWGn").click()
    
def scroll_page():
    """
        Objective:
            This function is used for scrolling by x-y pixels downwards.
        Input:
            None
        Output:
            page scrolls by x-y pixels
    """   
    driver.execute_script('window.scrollBy(0,500);') 

def insta_like(search_item,max):
    """
        Objective:
            This function is used for liking first max posts of search_item IG page.
        Input:
            search_item:sting
            max:integer
        Output:
            search_item IG page's first max post is liked.
    """
    insta_openProfile(search_item)
    
    #scrolling by 500 pixels downwards
    scroll_page()
    
    #wait until first post is clickable
    wait=WebDriverWait(driver,20) 
    first_pic= wait.until(EC.element_to_be_clickable((By.CLASS_NAME,'v1Nh3')))
    
    #opening first post
    first_pic=first_pic.click()
    i=1
    while i<=max :
        #wait until like button is clickable
        wait=WebDriverWait(driver,20) 
        like= wait.until(EC.element_to_be_clickable((By.XPATH,'//div[(@class="eo2As ")]/section/span/button')))
        
        #check if post is already liked
        if 'Like' not in like.get_attribute('innerHTML'):
            print('Already Liked for',i,'th post')
        else:
            like.click()
        i+=1
        
        #clicking right arrow for next post
        click_right_arrow()
        
    #closing current post
    click_close_post()
```

```python
insta_like('dilsefoodie',30)
```

```python
def insta_unlike(search_item,max):
    """
        Objective:
            This function is used for unliking first max posts of search_item IG page.
        Input:
            search_item:sting
            max:integer
        Output:
            search_item IG page's first max post is unliked.
    """
    insta_openProfile(search_item)
    
    #scrolling by 500 pixels downwards
    scroll_page()
      
    #wait until first post is clickable
    wait=WebDriverWait(driver,20) 
    first_pic= wait.until(EC.element_to_be_clickable((By.CLASS_NAME,'v1Nh3')))
    
    #opening first post
    first_pic=first_pic.click()
    i=1
    while i<=max :
        #wait until unlike button is clickable
        wait=WebDriverWait(driver,20) 
        unlike= wait.until(EC.element_to_be_clickable((By.XPATH,'//div[(@class="eo2As ")]/section/span/button')))
        
        #check if post is already unliked
        if 'Unlike' not in unlike.get_attribute('innerHTML'):
            print('Already Unliked for',i,'th post')
        else:
            unlike.click()
        i+=1
        
        #clicking right arrow for next post
        click_right_arrow()
        
    #closing current post
    click_close_post()
```

Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.
Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.

```python
def close_dialog_box():
    """
        Objective:
            This function is used for closing dialog box.
        Input:
            none
        Output:
            dialog box is closed.
    """
    close_dialog_box=driver.find_element_by_xpath('//span[(@aria-label="Close")]')  
    close_dialog_box.click()
    
def open_followerlist(search_item):
    """
        Objective:
            This function is used for opening followers list of search_item IG page.
        Input:
            search_item:sting
        Output:
            search_item IG page's followers list opens up.
    """
    insta_openProfile(search_item)
    open_followers=driver.find_element_by_xpath('//a[(@class="-nal3 ")]')  
    open_followers.click()

def lst_li_to_href(lst):
    """
        Objective:
           This function is used for creating list of href from given list.
        Input:
            lst:lst
        Output:
            returns list with hrefs.
    """
    href_lst=[]
    for user in lst:
        userLink = user.find_element_by_css_selector('a').get_attribute('href')
        href_lst.append(userLink)
    return href_lst
    
def insta_User_followerlist(search_item,max):
    """
        Objective:
            This function is used for extracting followers list of search_item IG page.
        Input:
            search_item:sting
            max:integer
        Output:
            Returns the list of followers of length max of search_item IG page's.
    """
    open_followerlist(search_item)
    
    #wait until dialog box of followers is located
    wait=WebDriverWait(driver,20) 
    followersList= wait.until(EC.presence_of_element_located((By.XPATH,'//div[(@role="dialog")]/div[2]/ul/div')))
    
    #extracting the length of followersList
    numberOfFollowersInList = len(followersList.find_elements_by_css_selector('li'))
    followersList.click()
    
    #for scrolling dialogbox
    actionChain = webdriver.ActionChains(driver)
    j=0
    while (numberOfFollowersInList < max):
        #performing pressing space bar on keyboard(key_down) and unpressing it(key_up)
        actionChain.key_down(Keys.SPACE).key_up(Keys.SPACE).perform()
        time.sleep(7)
        
        #extracting the length of followersList
        numberOfFollowersInList = len(followersList.find_elements_by_css_selector('li'))
        j+=1
        if j==1:
            followersList= wait.until(EC.presence_of_element_located((By.XPATH,'//div[(@role="dialog")]//div[(@class="isgrP")]/ul/div')))
            followersList.click()
            
    #creating list of followers href of length max 
    followers = lst_li_to_href(followersList.find_elements_by_css_selector('li')[:max])
        
    #close dialog box
    close_dialog_box()
    
    #returning list of followers
    return(followers)

def print_user_name_from_url(lst):
    """
        Objective:
            This function is used for printing the followers username from given url.
        Input:
            lst:list of strings
        Output:
            Prints the username from the list of urls.
    """
    for i in lst:
        print(i.split('/')[3])  
```

```python
def open_mutual_list(search_item):
    """
        Objective:
            This function is used for opening Mutuals list of search_item IG page.
        Input:
            search_item:sting
        Output:
            search_item IG page's mutuals list opens up.
    """
    insta_openProfile(search_item)
    
    #opening mutual list
    try:
        open_mutual_list=driver.find_element_by_xpath('//a[(@class="_32eiM")]')  
        open_mutual_list.click()
        #wait until link for open full mutual list is located
        wait=WebDriverWait(driver,10) 
        mutualListOpen= wait.until(EC.presence_of_element_located((By.XPATH,'//div[(@class="SfPzb")]/a')))
        mutualListOpen.click()
        return(1)
    #if no mutuals exists
    except NoSuchElementException:
        return(0)
    

def find_max_mutual(search_item):
    """
        Objective:
            This function is used for retrieving count of mutuals that are followered by the user .
        Input:
            search_item:string
        Output:
            Return the count of mutuals that are followered by the user.
    """
    toss=open_mutual_list(search_item)
    if toss:
        findmax=driver.find_element_by_xpath('//a[(@class="_32eiM")]/span')
        x=findmax.text
        x=x.split('+')[-1]
        #using regular expression to extract digits present in string
        x=float(re.findall(r'\d+',x)[0])
        max=int(x)+int(findmax.get_attribute('innerHTML').count('<span'))
        return max
    else:
        return(0)
    
def insta_User_mutual_list(search_item):
    """
        Objective:
            This function is used for extracting mutuals list of search_item IG page.
        Input:
            search_item:sting
        Output:
            Returns the list of mutuals of length max of search_item IG page's.
    """
    max=find_max_mutual(search_item)
    if max:
    
        #wait until dialog box of mutuals is located
        wait=WebDriverWait(driver,10) 
        mutualList= wait.until(EC.presence_of_element_located((By.XPATH,'//div[(@role="dialog")]/div[2]/ul/div')))

        #extracting the length of mutualsList
        numberOfmutualInList = len(mutualList.find_elements_by_css_selector('li'))
        mutualList.click()

        #for scrolling dialogbox
        actionChain = webdriver.ActionChains(driver)
        j=0
        while (numberOfmutualInList < max):
            #performing pressing space bar on keyboard(key_down) and unpressing it(key_up)
            actionChain.key_down(Keys.SPACE).key_up(Keys.SPACE).perform()
            time.sleep(5)

            #extracting the length of followersList
            numberOfmutualInList = len(mutualList.find_elements_by_css_selector('li'))
            j+=1
            if j==1:
                mutualList= wait.until(EC.presence_of_element_located((By.XPATH,'//div[(@role="dialog")]//div[(@class="isgrP")]/ul/div')))
                mutualList.click()

        #creating the list of mutuals href of length max(ie. whole)
        mutuals = lst_li_to_href(mutualList.find_elements_by_css_selector('li')[:max])

        #close dialog box
        close_dialog_box()

        #returning list of mutuals
        return(mutuals)
    else:
        return([])

def find_int_from_string(x):
    """
        Objective:
            This function is used for coverting string to integer value.
        Input:
            x:string
        Output:
            Returns the int value x.
    """
    if 'k' in x:
        y=x.split('k')[0]
        x=float(y)*1000
    elif 'm' in x:
        y=x.split('m')[0]
        x=float(y)*1000000
    else:
        if ',' in x:
            x=x.replace(',','')
        x=float(x)
    return int(x)
    """if ',' in x:
        x=x.replace(',','')
        #print(x)
    x=float(x)"""
    
def check_user_only_Follow(mutuals,username):
    """
        Objective:
            This function is used for checking if in mutuals list of search_item IG page, mutuals don't follow user.
        Input:
            foodtalkindia_mutuals:string
        Output:
            Returns the list of mutuals who don't follow user.
    """
    Link_Mfollow = []
    for i in mutuals:
        driver.get(i)
        findmax=driver.find_element_by_xpath('//section[(@class="zwlfE")]/ul/li[3]/a/span')
        x=findmax.text
        max=find_int_from_string(x)
        check_me=driver.find_element_by_xpath('//section[(@class="zwlfE")]/ul/li[3]')
        check_me.click()
        
        #wait until dialog box of followings is located
        wait=WebDriverWait(driver,20) 
        followingList= wait.until(EC.presence_of_element_located((By.XPATH,'//div[(@role="dialog")]//div[(@class="isgrP")]/ul/div')))
        
        #number of followings extracted
        numberOffollowingsInList = len(followingList.find_elements_by_css_selector('li'))
        followingList.click()
        
        #for scrolling dialogbox
        actionChain = webdriver.ActionChains(driver) 
        j=0
        while (numberOffollowingsInList < max-1):
            #performing pressing space bar on keyboard(key_down) and unpressing it(key_up)
            actionChain.key_down(Keys.SPACE).key_up(Keys.SPACE).perform()
            time.sleep(10)
            
            #number of followings extracted
            numberOffollowingsInList = len(followingList.find_elements_by_css_selector('li'))
            j+=1
            if j==1:
                followingList= wait.until(EC.presence_of_element_located((By.XPATH,'//div[(@role="dialog")]//div[(@class="isgrP")]/ul/div')))
                followingList.click()
                
        #creating the list of Link_Mfollow which includes those mutuals href that dont follow user.
        for user in followingList.find_elements_by_css_selector('li'):
            userLink = user.find_element_by_css_selector('a').get_attribute('href')
            if username in userLink:
                break
        else:
            Link_Mfollow.append(i)
            
        #close dialog box
        close_dialog_box()
        
    #return Links of Mutuals that dont follow the user
    return Link_Mfollow
```

## Q7 Check the story of ‘coding.ninjas’. Consider the following Scenarios and print error messages accordingly -
If You have already seen the story.
Or The user has no story.
Or View the story if not yet seen.

```python
def insta_checkStory(search_item):
    """
        Objective:
            This function is used for checking story of search_item IG page.
        Input:
            search_item:sting
        Output:
            Return viewing, already seen or no story available.
    """
    insta_openProfile(search_item)
    
    #finding the height of button(image)
    check=driver.find_element_by_xpath('//div[(@role="button")]')  
    canvas=check.get_attribute('innerHTML')
    
    #check if height=168, it may have an unviewed story or no story
    if 'height="168"' in canvas:
        try:
            check.click()
            #wait until story opens up of mutuals is located
            wait=WebDriverWait(driver,5) 
            story= wait.until(EC.presence_of_element_located((By.XPATH,'//div[(@class="yS4wN ")]')))
            story.click()
            print('Viewing')
        except TimeoutException:
            print('No Story')
            
    #else if height=166, story is seen
    elif 'height="166"' in canvas:
        print('Already Seen')
```
